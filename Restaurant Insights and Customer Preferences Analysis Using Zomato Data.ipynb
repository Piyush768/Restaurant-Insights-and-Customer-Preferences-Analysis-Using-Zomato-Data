```python
ratings= df2.groupby(['Aggregate rating','Rating text','Rating color']).size().reset_index().rename(columns= {0:'Rating_counts'})
```

```python
ratings
```

```python
import matplotlib as m
import matplotlib.pyplot as plt
import seaborn as sns

m.rcParams['figure.figsize'] = (12, 12)  # Increase figure size

# Fix: Remove the x and y arguments, and let 'data' handle them.
sns.barplot(x='Aggregate rating', y="Rating_counts", data=ratings)
plt.show()
```

This histogram shows the distribution of aggregate ratings for a dataset, where the x-axis represents the rating values and the y-axis represents the count of ratings. The data reveals a large concentration of ratings at 0, indicating a high number of unrated or zero-rated entries. After the 0 rating, most ratings are distributed normally, peaking around 3.3 to 3.4. There is a gradual decrease in the count of ratings as the rating value approaches 5. This suggests that while a few entries are highly rated, the majority of rated entries fall in the mid-range (around 3 to 3.5), indicating an average satisfaction level.

```python
m.rcParams['figure.figsize']=(12,12) # is to increase figure size
# Fix: Pass 'x', 'y', and 'hue' as column names, with 'data' as the DataFrame.
sns.barplot(x='Aggregate rating', y='Rating_counts', hue='Rating color', palette=['black','red','orange','yellow','green','#013220'], data=ratings)
plt.show()
```

The black bar at 0.0 represents a large number of unrated restaurants. The ratings from 2.0 to 4.9 show a clear distribution, with most restaurants clustered between 2.5 and 4.0. Lower ratings are represented in red, orange, and yellow, while higher ratings appear in green, with dark green indicating the highest ratings. This pattern suggests that the majority of restaurants have moderate to good ratings, with fewer achieving excellent ratings.

```python
sns.countplot(x='Rating color', palette=['black','red','orange','yellow','green','#013220'] , data= ratings)
```

This chart shows the distribution of Zomato ratings by color, where each color represents a rating range. **Orange** (below-average ratings) has the highest count, followed by **red** (low ratings) and **yellow** (average ratings). **Green** and **dark green** (good to excellent ratings) are less frequent, indicating fewer high-rated listings. **White** (unrated or very low-rated entries) has the lowest count. Overall, most listings have below-average or average ratings, suggesting a general perception of moderate satisfaction.

```python
city=df2['City'].value_counts().index
```

```python
plt.pie(city_values[:5],labels=city[:5],autopct="%.3f%%")
```

This pie chart displays the distribution of restaurant listings across various cities in the Delhi NCR region. **New Delhi** dominates with **68.87%** of listings, indicating it has the largest share of restaurants in the region. **Gurgaon** follows with **14.07%**, and **Noida** with **13.59%**. Smaller portions are represented by **Faridabad** (3.16%) and **Ghaziabad** (0.31%). This distribution shows that the majority of Zomato's restaurant listings in NCR are concentrated in New Delhi, with surrounding cities having a significantly smaller presence.

# Most order selling Cuisines on Zomato

```python
cuisines_index= df2['Cuisines'].value_counts().index
```

```python
cuisines_index
```

```python
cuisines_values= df2['Cuisines'].value_counts().values
```

```python
cuisines_values
```

```python
plt.pie(cuisines_values[:11],labels=cuisines_index[:11],autopct="%.3f%%")
```

This pie chart illustrates the distribution of popular cuisine types in the dataset. **North Indian** cuisine holds the largest share, accounting for **25.81%** of restaurant listings. **Chinese** and **Fast Food** cuisines follow, each with **9.67%**, reflecting a strong demand for these types. Other significant categories include **North Indian, Mughlai** (9.12%), **Cafe** (8.17%), and **Bakery** (5.95%). Specialized combinations like **North Indian, Chinese** (13.95%) and **North Indian, Mughlai, Chinese** (4.64%) indicate a preference for multi-cuisine offerings. This insight shows that North Indian cuisine is highly popular, often paired with other cuisines, indicating customer preference for variety within this category.
